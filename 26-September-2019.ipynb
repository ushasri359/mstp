```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.subplot(1,2,2)
linear_data=np.array([1,2,3,4,5,6,7,8])
plt.plot(linear_data,'-o')
```

```python
left=np.array([1,3,5])
right=np.array([2,4,6])
plt.figure()
plt.subplot(1,2,1)
plt.plot(left,'-o')

plt.subplot(1,2,2)
plt.plot(right,'-o')
```

```python
left=np.array([1,3,5])
right=np.array([2,4,6])
plt.figure()
plt.subplot(2,2,1)
plt.plot(left,'-o')

plt.subplot(2,2,2)
plt.plot(right,'-o')

plt.subplot(2,2,3)
plt.plot(right,'-o')

plt.subplot(2,2,4)
plt.plot(right,'-o')
```

```python
plt.figure()
ax1=plt.subplot(1,2,1)
plt.plot([1,2,3],np.array([50,60,70]),'-o',label='mid1')
plt.legend()
ax2=plt.subplot(1,2,2)
plt.plot([1,2,3],np.array([50,30,40]),'-x',label='mid2')
plt.legend()
```

```python
plt.figure()
ax1=plt.subplot(1,2,1)
plt.plot([1,2,3],np.array([50,60,70]),'-o')

ax2=plt.subplot(1,2,2,sharey=ax1)
plt.plot([1,2,3],np.array([50,30,40]),'-x')
```

### bar graph:

```python
plt.figure()
linear_data=np.array([10,20,30,40])
xvals=range(len(linear_data)) #[0,1,2,3]
plt.bar(xvals,linear_data,width=0.3)
```

```python
plt.figure()
linear_data=np.array([10,20,30,40])
xvals=range(len(linear_data)) #[0,1,2,3]
plt.barh(xvals,linear_data,height=0.3)      # barh,height
```

```python
import matplotlib.animation as animation
n=100
x=np.random.randn(n)
def update(curr):
    if curr==n:
        a.event_source.stop()
    plt.cla()
    bins=np.arange(-4,4,0.5)
    plt.hist(x[:curr],bins=bins)
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sampling the normal Distributiion')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n={}'.format(curr),[3,27])
```

```python
fig=plt.figure()
a=animation.FuncAnimation(fig,update,interval=100)
```
