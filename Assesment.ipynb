#### 2.Find the max,min and mean in values of sepalength,petallength,sepalwidth and petalwidth

```python
print('sepalwidth')
a=df['sepal_width'].max(),df['sepal_width'].min(),df['sepal_width'].mean()
print(a)
print('sepallength')
b=df['sepal_length'].max(),df['sepal_length'].min(),df['sepal_length'].mean()
print(b)
print('petalwidth')
c=df['petal_width'].max(),df['petal_width'].min(),df['petal_width'].mean()
print(c)
print('petallength')
d=df['petal_length'].max(),df['petal_length'].min(),df['petal_length'].mean()
print(d)
```
